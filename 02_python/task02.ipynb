```python
# int 以外は入力されないものとする
price = int(input("税抜価格を入力してください: "))

tax = 1.1       # 消費税
ship_fee = 350  # 送料

# 消費税計算は切り捨てとする
if price >= 2000:
    include_price = int(price * tax)
    print("送料は無料です")
    print(f"送料込税込価格は{include_price}円です")
elif price > 0:
    include_price = int(price * tax) + ship_fee
    print(f"送料は{ship_fee}円です")
    print(f"送料込税込価格は{include_price}円です")
else:
    print("税抜価格が不正です")
```
