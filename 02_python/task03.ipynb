```python
# int 以外は入力されないものとする
principal = int(input("元金を入力してください: "))
# int,float 以外は入力されないものとする
interest = float(input("年利を入力してください: "))

years = 10         # 年数
total = principal  # 総額 

# 年利 換算
interest_f = 1 + (interest / 100)  # 5% => 1.05

i = 0
while i < years:
    total *= interest_f
    i += 1

print(f"{years}年後の金額: {total}")
```
