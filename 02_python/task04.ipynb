```python
try:
    num1 = float(input("数値1を入力してください: "))
    num2 = float(input("数値2を入力してください: "))
    result = num1 / num2
except ZeroDivisionError:
    print("エラー: 0で割ることはできません")
except ValueError:
    print("エラー: 数値を入力してください")
except Exception as e:
    print(f"予期しないエラーが発生しました: {e}")
else:
    print(f"計算結果: {result}")
```
