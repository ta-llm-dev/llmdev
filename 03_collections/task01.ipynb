```python
grades = [85, 90, 78, 92, 88]

# 平均点
grade_average = sum(grades) / len(grades)
# 最高点
grade_max = max(grades)
# 最低点
grade_min = min(grades)
# 平均以上の学生
above_grade_average = [i for i in grades if i >= grade_average]

print(f"平均点: {grade_average}")
print(f"最高点: {grade_max}")
print(f"最低点: {grade_min}")
print(f"平均以上の学生数: {len(above_grade_average)}")
```
