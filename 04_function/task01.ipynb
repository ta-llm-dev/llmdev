```python
# 有効なユーザー名かどうか判定
def valid_name(name):
    valid = True

    if len(name) < 5 or len(name) > 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        valid = False
    
    if " " in name:
        print("エラー: ユーザー名に空白を含めることはできません")
        valid = False
    
    if not name.isalnum():
        print("エラー: ユーザー名は英数字のみ使用できます")
        valid = False
    
    return valid


input_name = input("ユーザー名を入力してください: ")

if valid_name(input_name):
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")


```
